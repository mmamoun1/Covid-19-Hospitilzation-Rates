```python
# First Step: I will import pandas to use the dataframe functionality
import pandas as pd
# Now, I can import the dataset into a dataframe named df.
df = pd.read_csv('https://healthdata.gov/resource/6xf2-c3ie.csv')
# Now, I can check the shape of the dataframe to see if it has the expected 72.6k rows of data'
df.shape
```
